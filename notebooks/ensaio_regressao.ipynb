## 0.1 Imports

```python
import pandas as pd
import numpy as np

from sklearn import metrics as mt
from sklearn import tree as tr
from sklearn import ensemble as en
from sklearn import linear_model as lm
from sklearn import preprocessing as pp
```

## 0.2 Load dataset

```python
X_train = pd.read_csv('X_training.csv')
print(X_train.shape)
X_train.head()
```

```python
y_train = pd.read_csv('y_training.csv')
print(y_train.shape)
y_train.head()
```

```python
# validação
X_val = pd.read_csv('X_validation.csv')
print(X_val.shape)
X_val.head()
```

```python
y_val = pd.read_csv('y_val.csv')
print(y_val.shape)
y_val.head()
```

```python
# Teste
X_test = pd.read_csv('X_test.csv')
print(X_test.shape)
X_test.head()
```

```python
y_test = pd.read_csv('y_test.csv')
print(y_test.shape)
y_test.head()
```

# 1.0 Decision Tree Regressor

## 1.1 Treino

```python
depth = np.arange(10, 21, 1)
metricas = []
for d in depth:
    model = tr.DecisionTreeRegressor(max_depth = d).fit(X_train, y_train)
    yhat = model.predict(X_train)
    #metricas
    r2 = round(mt.r2_score(y_train, yhat),3)
    mse = round(mt.mean_squared_error(y_train, yhat),3)
    rmse = round(np.sqrt(mse),3)
    mae = round(mt.mean_absolute_error(y_train, yhat),3)
    mape = round(mt.mean_absolute_percentage_error(y_train, yhat),3)

    m_value= {'max_depth': d, 'R2': r2, 'MSE': mse, 'RMSE': rmse, 'MAE': mae , 'MAPE': mape}
    metricas.append(m_value)

m = pd.DataFrame(metricas)
m.sort_values(by=['R2', 'MAE'], ascending=[False, True])
```

## 1.2 dados de validação

```python
depth = np.arange(1, 11, 1)
metricas = []
for d in depth:
    model = tr.DecisionTreeRegressor(max_depth = d).fit(X_train, y_train)
    yhat = model.predict(X_val)
    #metricas
    r2 = round(mt.r2_score(y_val, yhat),3)
    mse = round(mt.mean_squared_error(y_val, yhat),3)
    rmse = round(np.sqrt(mse),3)
    mae = round(mt.mean_absolute_error(y_val, yhat),3)
    mape = round(mt.mean_absolute_percentage_error(y_val, yhat),3)

    m_value= {'max_depth': d, 'R2': r2, 'MSE': mse, 'RMSE': rmse, 'MAE': mae , 'MAPE': mape}
    metricas.append(m_value)

m = pd.DataFrame(metricas)
m.sort_values(by=['R2', 'MAE'], ascending=[False, True])
```

## 1.3 Dados de teste

```python
d=5
metricas = []


model = tr.DecisionTreeRegressor(max_depth = d)
model.fit(pd.concat([X_train, X_val]), pd.concat([y_train, y_val]))
yhat = model.predict(X_test)
#metricas
r2 = round(mt.r2_score(y_test, yhat),3)
mse = round(mt.mean_squared_error(y_test, yhat),3)
rmse = round(np.sqrt(mse),3)
mae = round(mt.mean_absolute_error(y_test, yhat),3)
mape = round(mt.mean_absolute_percentage_error(y_test, yhat),3)

m_value= {'max_depth': d, 'R2': r2, 'MSE': mse, 'RMSE': rmse, 'MAE': mae , 'MAPE': mape}
metricas.append(m_value)

m = pd.DataFrame(metricas)
m.sort_values(by=['R2', 'MAE'], ascending=[False, True])
```
