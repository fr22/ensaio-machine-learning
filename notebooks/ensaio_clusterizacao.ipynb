# 0.0 Imports

```python
import pandas as pd
import numpy as np
from sklearn import cluster as c
from sklearn import metrics as mt
```

## 0.1 Load dataset

```python
X = pd.read_csv('X_dataset.csv')
print(X.shape)
X.head()
```

## 1.0 Kmeans

```python
clusters = np.arange(2, 11, 1)
metricas = []

for n in clusters:
    model = c.KMeans(n_clusters = n)
    labels = model.fit_predict(X)
    #metricas
    ss = round( mt.silhouette_score(X, labels), 3)
    m_value = {'n_clusters': n, 'Silhouette_Score': ss}
    metricas.append(m_value)

m = pd.DataFrame(metricas)
m.sort_values(by='Silhouette_Score', ascending=False)

```

## 2.0 Affinity Propagation

```python
#pref = np.arange(-100, -40, 1)
p = -50
damping = [0.6, 0.7, 0.8, 0.9]
metricas = []

for d in damping:
    model = c.AffinityPropagation(preference = p, damping = d, max_iter = 500 )
    model.fit(X)
    
    labels = model.predict(X)
    #metricas
    ss = round( mt.silhouette_score(X, labels), 3)
    n = len(np.unique(labels))
    m_value = {'preference': p, 'damping': d, 'n_clusters': n, 'Silhouette_Score': ss}
    metricas.append(m_value) 

m = pd.DataFrame(metricas)
m.sort_values(by='Silhouette_Score', ascending=False)
```
