# 0.1 Imports

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics as mt
from sklearn import tree as tr
from sklearn import ensemble as en
from sklearn import linear_model as lm
```

```python
# Load dataset
pd.set_option('display.max_columns', None)
X_train = pd.read_csv('X_training.csv')

X_train.head()
```

```python
# removendo id
X_train = X_train.drop(X_train.columns[0], axis=1)
X_train.head(1)
```

```python
X_train.shape
```

```python
X_train_s = X_train.sample(frac=0.1, random_state=0)
X_train_s.shape
```

```python
y_train = pd.read_csv('y_training.csv')

y_train= np.ravel(y_train)
y_train.shape
```

```python
y_train_s = y_train[X_train_s.index]
y_train_s.shape
```

```python
#X_val
X_val = pd.read_csv('X_validation.csv')
X_val_s = X_val.sample(frac=0.1, random_state=0)
X_val_s.shape
```

```python
#removendo id
X_val_s = X_val_s.drop(X_val_s.columns[0], axis=1)
X_val_s.head(1)
```

```python
# y_val
y_val = pd.read_csv('y_validation.csv')
y_val = np.ravel(y_val)
y_val_s = y_val[X_val_s.index]
y_val_s.shape
```

```python
#X_test
X_test = pd.read_csv('X_test.csv')
X_test_s = X_test.sample(frac=0.1, random_state=0)
X_test_s.shape

X_test_s = X_test_s.drop(X_test_s.columns[0], axis=1)
X_test_s.head(1)
```

```python
X_test_s.shape
```

```python
#y_test
y_test = pd.read_csv('y_test.csv')
y_test = np.ravel(y_test)
y_test_s = y_test[X_test_s.index]
y_test_s.shape
```

# 1.0 Avaliação KNN

## 1.1 Dados de treino

```python
# Instanciando
param = np.arange(1,11, 1)
for p in param:
    model = KNeighborsClassifier(n_neighbors=p).fit(X_train_s,y_train_s)
    #predict
    yhat = model.predict(X_train_s)

    # metricas
    acc = round(mt.accuracy_score(y_train_s, yhat),3)
    pr = round(mt.precision_score(y_train_s,yhat),3)
    rc = round(mt.recall_score(y_train_s,yhat),3)
    f1 = round(mt.f1_score(y_train_s,yhat),3)

    print(f'Valor n_neighbor: {p}')
    print(f'Métricas. Acuracia: {acc}, Precision: {pr}, Recall: {rc}, F1_score: {f1}')
    print('=========================================')
    
```

## 1.2 Dados validação

```python
# Instanciando
param = np.arange(1,11, 1)
for p in param:
    model = KNeighborsClassifier(n_neighbors=p).fit(X_train_s,y_train_s)
    #predict
    yhat = model.predict(X_val_s)

    # metricas
    acc = round(mt.accuracy_score(y_val_s, yhat),3)
    pr = round(mt.precision_score(y_val_s,yhat),3)
    rc = round(mt.recall_score(y_val_s,yhat),3)
    f1 = round(mt.f1_score(y_val_s,yhat),3)

    print(f'Valor n_neighbor: {p}')
    print(f'Métricas. Acuracia: {acc}, Precision: {pr}, Recall: {rc}, F1_score: {f1}')
    print('=========================================')
    
```

## 1.3 Dados de teste

```python
p=7
model = KNeighborsClassifier(n_neighbors=p).fit(X_train_s,y_train_s)
#predict
yhat = model.predict(X_test_s)

# metricas
acc = round(mt.accuracy_score(y_test_s, yhat),3)
pr = round(mt.precision_score(y_test_s,yhat),3)
rc = round(mt.recall_score(y_test_s,yhat),3)
f1 = round(mt.f1_score(y_test_s,yhat),3)

print(f'Valor n_neighbor: {p}')
print(f'Métricas. Acuracia: {acc}, Precision: {pr}, Recall: {rc}, F1_score: {f1}')
print('=========================================')
```

# 2.0 Avaliação Decision tree

## 2.1 Dados de treino

```python
param = np.arange(2, 11, 1)
metricas = []
for p in param:
    
    model = tr.DecisionTreeClassifier(max_depth=p).fit(X_train_s, y_train_s)
    yhat = model.predict(X_train_s)
    #metricas
    acc = round(mt.accuracy_score(y_train_s, yhat),3)
    pr = round(mt.precision_score(y_train_s,yhat),3)
    rc = round(mt.recall_score(y_train_s,yhat),3)
    f1 = round(mt.f1_score(y_train_s,yhat),3)
    m_value = {'max_depth': p, 'acuracia': acc, 'precision': pr, 'recall': rc, 'f1_score': f1 }
    metricas.append(m_value)

r = pd.DataFrame(metricas)
r
```

## 2.2 Dados de validação

```python
param = np.arange(2, 16, 1)
metricas = []
for p in param:
    
    model = tr.DecisionTreeClassifier(max_depth=p).fit(X_train_s, y_train_s)
    yhat = model.predict(X_val_s)
    #metricas
    acc = round(mt.accuracy_score(y_val_s, yhat),3)
    pr = round(mt.precision_score(y_val_s,yhat),3)
    rc = round(mt.recall_score(y_val_s,yhat),3)
    f1 = round(mt.f1_score(y_val_s,yhat),3)
    m_value = {'max_depth': p, 'acuracia': acc, 'precision': pr, 'recall': rc, 'f1_score': f1 }
    metricas.append(m_value)

r = pd.DataFrame(metricas)
r
```

# 3.0 Avaliação Random forest

## 3.1 Dados de treino

```python
#e = 100
#d = 10
metricas = []
estimators = np.arange(10, 101, 10)
depth = np.arange(5, 16, 1)
for e in estimators:
    for d in depth:
        
        model = en.RandomForestClassifier(n_estimators = e, max_depth = d).fit(X_train_s, y_train_s)
        yhat = model.predict(X_train_s)
        
        acc = round(mt.accuracy_score(y_train_s, yhat),3)
        pr = round(mt.precision_score(y_train_s,yhat),3)
        rc = round(mt.recall_score(y_train_s,yhat),3)
        f1 = round(mt.f1_score(y_train_s,yhat),3)

        m_value = {'n_estimators': e ,'max_depth': d, 'acuracia': acc, 'precision': pr, 'recall': rc, 'f1_score': f1 }
        metricas.append(m_value)
m = pd.DataFrame(metricas)
m
```

```python
m.sort_values(by=['acuracia', 'f1_score'], ascending=False)
```

# 4.0 Logistic Regression

## 4.1 Dados de treino

```python
C = np.arange(0.1, 1, 0.4)
solvers = ['lbfgs', 'newton-cg', 'newton-cholesky', 'sag']
max_iters = np.arange(90, 111, 10)
metricas=[]

for c in C:
    for s in solvers:
        for i in max_iters:
            
            model = lm.LogisticRegression(C = c, solver = s, max_iter = i).fit(X_train_s, y_train_s)
            yhat= model.predict(X_train_s)
            
            acc = round(mt.accuracy_score(y_train_s, yhat),3)
            pr = round(mt.precision_score(y_train_s,yhat),3)
            rc = round(mt.recall_score(y_train_s,yhat),3)
            f1 = round(mt.f1_score(y_train_s,yhat),3)
    
            m_value = {'C': c ,'solver': s, 'max_iter': i, 'acuracia': acc, 'precision': pr, 'recall': rc, 'f1_score': f1 }
            metricas.append(m_value)
m = pd.DataFrame(metricas)
m.sort_values(by=['acuracia', 'f1_score'], ascending=False)
```
